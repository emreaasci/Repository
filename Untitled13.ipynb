```python
import numpy as np
import os
from tensorflow.keras.preprocessing import image
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout, Activation, TimeDistributed, LSTM
from tensorflow.keras.optimizers import Adam
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split as sp
from tensorflow.keras.utils import to_categorical
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report, roc_auc_score
import matplotlib.pyplot as plt
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout, Activation, BatchNormalization, GlobalAveragePooling2D
from tensorflow.keras.optimizers import RMSprop
from sklearn.metrics import confusion_matrix, accuracy_score, roc_auc_score, recall_score, precision_score, f1_score

img_height = 100
img_width = 100

positive_path = "/content/gdrive/MyDrive/veri/pozitif"
negative_path = "/content/gdrive/MyDrive/veri/negatif"

def image_to_vector(img_file_path):
    img = image.load_img(img_file_path, target_size=(img_height, img_width))
    x = image.img_to_array(img)
    return x

def get_images(path, label):
    image_list = []
    class_list = []
    for _, _, filenames in os.walk(path):
        for filename in filenames:
            img = image_to_vector(os.path.join(path, filename))
            image_list.append(img)
            class_list.append(label)
    return image_list, class_list

def load_data():
    positive_images, positive_labels = get_images(positive_path, "pozitif")
    negative_images, negative_labels = get_images(negative_path, "negatif")

    all_images = np.array(positive_images + negative_images)
    all_labels = np.array(positive_labels + negative_labels)

    encoder = LabelEncoder()
    classes = encoder.fit_transform(all_labels)
    classes = to_categorical(classes, num_classes=2)

    return all_images, classes

def create_cnn_model():
    model = Sequential()

    model.add(Conv2D(64, (3, 3), input_shape=(img_height, img_width, 3)))
    model.add(Activation('relu'))
    model.add(MaxPooling2D(pool_size=(2, 2)))

    model.add(TimeDistributed(Flatten()))
    model.add(LSTM(units=100, kernel_initializer='constant'))

    model.add(Dropout(0.5))
    model.add(Dense(2))  # İki sınıf için
    model.add(Activation('softmax'))

    model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
    return model

# Veri yükleme ve model eğitimi
images, labels = load_data()
images = images / 255.0
x_train, x_test, y_train, y_test = sp(images, labels, test_size=0.2, random_state=42)

model = create_cnn_model()

# Store the training history for plotting
history = model.fit(x_train, y_train, epochs=10, validation_data=(x_test, y_test))

# Plot training history
plt.figure(figsize=(12, 6))

# Plot training & validation accuracy values
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('Model accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend(['Train', 'Test'], loc='upper left')

# Plot training & validation loss values
plt.subplot(1, 2, 2)
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend(['Train', 'Test'], loc='upper left')

# Show plots
plt.tight_layout()
plt.show()

# Evaluation metrics
y_pred = model.predict(x_test)
conf_mat = confusion_matrix(y_test.argmax(axis=1), y_pred.argmax(axis=1))
acc = accuracy_score(y_test.argmax(axis=1), y_pred.argmax(axis=1))
roc_auc = roc_auc_score(y_test, y_pred, multi_class="ovr")
sensitivity = recall_score(y_test.argmax(axis=1), y_pred.argmax(axis=1), average='weighted')
specificity = (conf_mat[0, 0] / (conf_mat[0, 0] + conf_mat[0, 1]))  # True Negative / (True Negative + False Positive)
precision = precision_score(y_test.argmax(axis=1), y_pred.argmax(axis=1), average='weighted')
f_score = f1_score(y_test.argmax(axis=1), y_pred.argmax(axis=1), average='weighted')

print('Accuracy Score: ', acc)
print('Roc_AUC Score: ', roc_auc)
print('Sensitivity: ', sensitivity)
print('Specificity: ', specificity)
print('Precision: ', precision)
print('F-Score: ', f_score)
print(conf_mat)

```

```python
# Gerekli kütüphaneleri içe aktarma
import numpy as np  # Sayısal işlemler için
import os  # Dosya ve dizin işlemleri için
from tensorflow.keras.preprocessing import image  # Görüntü işleme için
from tensorflow.keras.models import Sequential  # Model oluşturma için
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout, Activation, TimeDistributed, LSTM  # CNN ve LSTM katmanları için
from tensorflow.keras.optimizers import Adam  # Optimizasyon algoritması için
from sklearn.preprocessing import LabelEncoder  # Etiket kodlaması için
from sklearn.model_selection import train_test_split as sp  # Veri setini bölme işlemi için
from tensorflow.keras.utils import to_categorical  # Etiketleri kategorik formata dönüştürme için
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report, roc_auc_score  # Performans metrikleri için
import matplotlib.pyplot as plt  # Grafik çizdirme için

# Görüntü boyutlarını tanımlama
img_height = 100
img_width = 100

# Pozitif ve negatif görüntülerin bulunduğu dizin yolları
positive_path = "/content/gdrive/MyDrive/veri/pozitif"
negative_path = "/content/gdrive/MyDrive/veri/negatif"

# Görüntüyü vektöre dönüştürme fonksiyonu
def image_to_vector(img_file_path):
    img = image.load_img(img_file_path, target_size=(img_height, img_width))  # Görüntüyü yükle ve yeniden boyutlandır
    x = image.img_to_array(img)  # Görüntüyü diziye çevir
    return x

# Belirli bir yoldaki görüntüleri ve sınıflarını yükleme fonksiyonu
def get_images(path, label):
    image_list = []
    class_list = []
    for _, _, filenames in os.walk(path):  # Belirtilen dizindeki dosyaları yinele
        for filename in filenames:
            img = image_to_vector(os.path.join(path, filename))  # Her bir görüntüyü vektöre dönüştür
            image_list.append(img)
            class_list.append(label)
    return image_list, class_list

# Veri yükleme fonksiyonu
def load_data():
    positive_images, positive_labels = get_images(positive_path, "pozitif")  # Pozitif görüntüleri yükle
    negative_images, negative_labels = get_images(negative_path, "negatif")  # Negatif görüntüleri yükle

    all_images = np.array(positive_images + negative_images)  # Tüm görüntüleri birleştir
    all_labels = np.array(positive_labels + negative_labels)  # Tüm etiketleri birleştir

    encoder = LabelEncoder()  # Etiket kodlayıcı
    classes = encoder.fit_transform(all_labels)  # Etiketleri sayısal hale getir
    classes = to_categorical(classes, num_classes=2)  # Etiketleri kategorik formata dönüştür

    return all_images, classes

# CNN modeli oluşturma fonksiyonu
def create_cnn_model():
    model = Sequential()

    model.add(Conv2D(64, (3, 3), input_shape=(img_height, img_width, 3)))  # Evrişim katmanı
    model.add(Activation('relu'))  # Aktivasyon fonksiyonu
    model.add(MaxPooling2D(pool_size=(2, 2)))  # Havuzlama katmanı

    model.add(TimeDistributed(Flatten()))  # Zaman dağıtılmış düzleştirme
    model.add(LSTM(units=100, kernel_initializer='constant'))  # LSTM katmanı

    model.add(Dropout(0.5))  # Aşırı uyumu önlemek için Dropout
    model.add(Dense(2))  # Tam bağlantılı katman, çıktı sınıf sayısı
    model.add(Activation('softmax'))  # Çoklu sınıf sınıflandırma için softmax aktivasyonu

    model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])  # Modeli derle
    return model

# Veri yükleme ve model eğitimi
images, labels = load_data()
images = images / 255.0  # Görüntüleri normalize et
x_train, x_test, y_train, y_test = sp(images, labels, test_size=0.2, random_state=42)  # Veriyi eğitim ve test setlerine ayır

model = create_cnn_model()

# Eğitim geçmişini depolamak için
history = model.fit(x_train, y_train, epochs=10, validation_data=(x_test, y_test))  # Modeli eğit

# Eğitim geçmişini çizdirme
plt.figure(figsize=(12, 6))

# Eğitim ve doğrulama doğruluğu değerlerini çizdir
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('Model accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend(['Train', 'Test'], loc='upper left')

# Eğitim ve doğrulama kaybı değerlerini çizdir
plt.subplot(1, 2, 2)
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend(['Train', 'Test'], loc='upper left')

# Grafikleri göster
plt.tight_layout()
plt.show()

# Değerlendirme metrikleri
y_pred = model.predict(x_test)  # Test verisi üzerinde tahmin yap
conf_mat = confusion_matrix(y_test.argmax(axis=1), y_pred.argmax(axis=1))  # Karışıklık matrisi
acc = accuracy_score(y_test.argmax(axis=1), y_pred.argmax(axis=1))  # Doğruluk skoru
roc_auc = roc_auc_score(y_test, y_pred, multi_class="ovr")  # ROC-AUC skoru
sensitivity = recall_score(y_test.argmax(axis=1), y_pred.argmax(axis=1), average='weighted')  # Hassasiyet
specificity = (conf_mat[0, 0] / (conf_mat[0, 0] + conf_mat[0, 1]))  # Özgüllük
precision = precision_score(y_test.argmax(axis=1), y_pred.argmax(axis=1), average='weighted')  # Kesinlik
f_score = f1_score(y_test.argmax(axis=1), y_pred.argmax(axis=1), average='weighted')  # F1 skoru

# Metrikleri yazdır
print('Accuracy Score: ', acc)
print('Roc_AUC Score: ', roc_auc)
print('Sensitivity: ', sensitivity)
print('Specificity: ', specificity)
print('Precision: ', precision)
print('F-Score: ', f_score)
print(conf_mat)  # Karışıklık matrisini yazdır
```

```python
import numpy as np
import os
from tensorflow.keras.preprocessing import image
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout, Activation, BatchNormalization, GlobalAveragePooling2D
from tensorflow.keras.optimizers import RMSprop
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split as sp
from tensorflow.keras.utils import to_categorical
from sklearn.metrics import confusion_matrix, accuracy_score, roc_auc_score, recall_score, precision_score, f1_score
import matplotlib.pyplot as plt
from tensorflow.keras.applications import Xception
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Flatten, Dense

# Resimlerin boyutları
img_height = 100
img_width = 100

# Pozitif ve negatif resimlerin yolu
positive_path = "/content/gdrive/MyDrive/veri/pozitif"
negative_path = "/content/gdrive/MyDrive/veri/negatif"

# Resim dosyasını bir vektöre dönüştüren fonksiyon
def image_to_vector(img_file_path):
    img = image.load_img(img_file_path, target_size=(img_height, img_width))
    x = image.img_to_array(img)
    return x

# Verilen yoldaki resimleri ve etiketlerini döndüren fonksiyon
def get_images(path, label):
    image_list = []
    class_list = []
    for _, _, filenames in os.walk(path):
        for filename in filenames:
            img = image_to_vector(os.path.join(path, filename))
            image_list.append(img)
            class_list.append(label)
    return image_list, class_list

# Veriyi yükleyen fonksiyon
def load_data():
    positive_images, positive_labels = get_images(positive_path, "pozitif")
    negative_images, negative_labels = get_images(negative_path, "negatif")

    all_images = np.array(positive_images + negative_images)
    all_labels = np.array(positive_labels + negative_labels)

    # Etiketleri kodlayan ve one-hot encoding uygulayan LabelEncoder kullanılır
    encoder = LabelEncoder()
    classes = encoder.fit_transform(all_labels)
    classes = to_categorical(classes, num_classes=2)

    return all_images, classes

# Önceden eğitilmiş Xception modelini oluşturan fonksiyon
def pretrained_Xception():
    base_model = Xception(weights='imagenet', include_top=False, input_shape=(img_height, img_width, 3))
    base_model.trainable = False

    model = Sequential([
        base_model,
        Flatten(),
        Dense(256, activation='relu'),
        Dense(2, activation='softmax')
    ])

    model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
    return model

# Eğitim sonuçlarını çizen fonksiyon
def plot_metrics(history):
    plt.figure(figsize=(12, 4))

    # Eğitim ve doğrulama doğruluk değerlerini çiz
    plt.subplot(1, 2, 1)
    plt.plot(history.history['accuracy'])
    plt.plot(history.history['val_accuracy'])
    plt.title('Model Doğruluk Değerleri')
    plt.xlabel('Epoch')
    plt.ylabel('Doğruluk')
    plt.legend(['Eğitim', 'Test'], loc='upper left')

    # Eğitim ve doğrulama kayıp değerlerini çiz
    plt.subplot(1, 2, 2)
    plt.plot(history.history['loss'])
    plt.plot(history.history['val_loss'])
    plt.title('Model Kayıp Değerleri')
    plt.xlabel('Epoch')
    plt.ylabel('Kayıp')
    plt.legend(['Eğitim', 'Test'], loc='upper left')

    plt.tight_layout()
    plt.show()

# Veriyi yükleme ve modeli eğitme
images, labels = load_data()
images = images / 255.0
x_train, x_test, y_train, y_test = sp(images, labels, test_size=0.2, random_state=42)

model = pretrained_Xception()
history = model.fit(x_train, y_train, epochs=10, validation_data=(x_test, y_test))

# Doğruluk ve kayıp grafiğini çizme
plot_metrics(history)

# Değerlendirme metrikleri
y_pred = model.predict(x_test)
conf_mat = confusion_matrix(y_test.argmax(axis=1), y_pred.argmax(axis=1))
acc = accuracy_score(y_test.argmax(axis=1), y_pred.argmax(axis=1))
roc_auc = roc_auc_score(y_test, y_pred, multi_class="ovr")
sensitivity = recall_score(y_test.argmax(axis=1), y_pred.argmax(axis=1), average='weighted')
specificity = (conf_mat[0, 0] / (conf_mat[0, 0] + conf_mat[0, 1]))  # True Negative / (True Negative + False Positive)
precision = precision_score(y_test.argmax(axis=1), y_pred.argmax(axis=1), average='weighted')
f_score = f1_score(y_test.argmax(axis=1), y_pred.argmax(axis=1), average='weighted')

print('Doğruluk Skoru: ', acc)
print('Roc_AUC Skoru: ', roc_auc)
print('Hassasiyet: ', sensitivity)
print('Özgünlük: ', specificity)
print('Kesinlik: ', precision)
print('F-Skor: ', f_score)
print(conf_mat)
```
