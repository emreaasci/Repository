```python
import networkx as nx
import random
import matplotlib.pyplot as plt

# Ağınızı 'G' olarak yükleyin veya oluşturun
# Örneğin: G = nx.gnp_random_graph(30, 0.05)

def plot_resilience(G, iterations=None):
    # Başlangıçta ağın en büyük bağlantılı bileşeninin boyutu
    initial_largest_cc = len(max(nx.connected_components(G), key=len))

    # Düğüm sayısı
    num_nodes = G.number_of_nodes()

    # Varsayılan iterasyon sayısı olarak düğüm sayısını kullan
    if iterations is None:
        iterations = num_nodes - 1

    # Her düğüm çıkarıldığında en büyük bağlantılı bileşenin boyutunu saklamak için liste
    sizes = [initial_largest_cc]

    for i in range(iterations):
        # Rastgele bir düğüm seç ve çıkar
        node_to_remove = random.choice(list(G.nodes()))
        G.remove_node(node_to_remove)

        # En büyük bağlantılı bileşenin yeni boyutunu hesapla ve kaydet
        if len(G) > 0:  # Ağda hala düğüm varsa
            largest_cc = len(max(nx.connected_components(G), key=len))
        else:
            largest_cc = 0
        sizes.append(largest_cc)

    # Yüzdelik olarak boyutları hesapla
    sizes_percent = [s / initial_largest_cc * 100 for s in sizes]

    # Grafiği çiz
    plt.plot(range(num_nodes - iterations, num_nodes + 1), sizes_percent)
    plt.xlabel('Çıkarılan Düğüm Sayısı')
    plt.ylabel('En Büyük Bağlantılı Bileşenin Boyutu (%)')
    plt.title('Ağ Dayanıklılığı')
    plt.show()

# Fonksiyonu çağırarak ağ dayanıklılığını çiz
plot_resilience(G.copy())

```

```python
import networkx as nx
import matplotlib.pyplot as plt
import random

# Grafiği oluştur
edges = [


    ('walt', 'fly'), ('walt', 'jesse'), ('walt', 'carmen'), ('walt', 'delcavoli'),
    ('walt', 'hank'), ('walt', 'gretchen'), ('walt', 'elliott'), ('walt', 'bogdan'),
    ('walt', 'skyler'), ('walt', 'holly'), ('walt', 'walt jr'), ('walt', 'hector'),
    ('walt', 'krazy-8'), ('walt', 'tuco'), ('walt', 'emilio'), ('walt', 'gus'),
    ('walt', 'gale'), ('walt', 'victor'), ('walt', 'mike'), ('walt', 'saul'),
    ('jesse', 'mike'), ('jesse', 'wendy'), ('jesse', 'group leader'), ('jesse', 'andrea'),
    ('jesse', 'brock'), ('jesse', 'tomas'), ('jesse', 'jane'), ('jesse', 'skinny pete'),
    ('jesse', 'combo'), ('jesse', 'badger'), ('jesse', 'clovis'), ('jesse', 'old joe'),
    ('jesse', 'mrs p'), ('jesse', 'mr p'), ('jesse', 'jake'), ('jesse', 'saul'),
    ('saul', 'mike'), ('saul', 'huell'), ('saul', 'kuby'), ('saul', 'francesca'), ('saul', 'ed'),
    ('group leader', 'andrea'), ('andrea', 'brock'), ('brock', 'tomas'), ('tomas', 'combo'),
    ('skinny pete', 'combo'), ('skinny pete', 'badger'), ('skinny pete', 'clovis'), ('skinny pete', 'old joe'),
    ('combo', 'badger'), ('combo', 'clovis'), ('combo', 'old joe'), ('badger', 'clovis'), ('badger', 'old joe'),
    ('clovis', 'old joe'), ('jane', 'donald'), ('mr p', 'mrs p'), ('mrs p', 'jake'), ('mr p', 'jake'),
    ('mike', 'victor'), ('mike', 'gus'), ('mike', 'tyrus'), ('mike', 'chris'), ('mike', 'dan'), ('mike', 'kaylee'),
    ('mike', 'todd'), ('mike', 'lydia'), ('mike', 'duane'), ('todd', 'declan'), ('todd', 'jack'), ('todd', 'lydia'),
    ('jack', 'kenny'), ('jack', 'frakie'), ('jack', 'lester'), ('jack', 'matt'), ('lydia', 'daclan'), ('lydia', 'schuler'),
    ('duane', 'peng'), ('gus', 'gale'), ('gus', 'victor'), ('gus', 'tyrus'), ('gus', 'cynthia'), ('gus', 'dennis'),
    ('gus', 'barry'), ('gus', 'max'), ('gus', 'hector'), ('gus', 'don eladio'), ('skyler', 'holly'), ('skyler', 'gretchen'),
    ('skyler', 'elliott'), ('skyler', 'bogdan'), ('skyler', 'marie'), ('skyler', 'hank'), ('skyler', 'ted'), ('skyler', 'pamela'),
    ('skyler', 'walt jr'), ('walt jr', 'holly'), ('walt jr', 'louis'), ('hank', 'scott'), ('hank', 'tim'), ('hank', 'janice'),
    ('hank', 'gomez'), ('hank', 'merkert'), ('hank', 'ramey'), ('hank', 'marie'), ('hank', 'tortuga'), ('hank', 'leonel'),
    ('hank', 'marco'), ('hank', 'hector'), ('hank', 'tuco'), ('hank', 'marie'), ('artic', 'scott'), ('tuco', 'no-doze'),
    ('tuco', 'gonzo'), ('tuco', 'hector'), ('tuco', 'no-doze'), ('don eladio', 'hector'), ('don eladio', 'max'),
    ('don eladio', 'marco'),
    ('don eladio', 'leonel'),
    ('don eladio', 'tortuga'),
    ('don eladio', 'juan'),
    ('don eladio', 'benicio'),
    ('don eladio', 'gaff'), ('krazy-8', 'emilio'), ('juan', 'tortuga'), ('hector', 'marco'), ('marco', 'leonel'),
    ('frakie', 'lester'), ('frakie', 'kenny'), ('frakie', 'matt'), ('lester', 'kenny'), ('lester', 'matt'),
    ('kenny', 'matt')


]
G = nx.Graph()
G.add_edges_from(edges)

def plot_network_resilience(G, title='Ağ Dayanıklılığı'):
    # Başlangıçta ağın en büyük bağlantılı bileşeninin boyutu
    initial_largest_cc = len(max(nx.connected_components(G), key=len))

    # Düğüm sayısı
    num_nodes = len(G.nodes())

    # En büyük bağlantılı bileşenin boyutlarını saklamak için liste
    sizes = [initial_largest_cc]
    largest_ccs = [initial_largest_cc]


    # Ağ kopyası
    G_copy = G.copy()

    while len(G_copy) > 1:
        # Rastgele bir düğüm seç ve çıkar
        node_to_remove = random.choice(list(G_copy.nodes()))
        G_copy.remove_node(node_to_remove)

        # En büyük bağlantılı bileşenin yeni boyutunu hesapla ve kaydet
        largest_cc = len(max(nx.connected_components(G_copy), key=len))
        sizes.append(largest_cc)

    # Grafiği çiz
    plt.plot(range(num_nodes, num_nodes - len(sizes), -1), sizes)
    plt.xlabel('Kalan Düğüm Sayısı')
    plt.ylabel('En Büyük Bağlantılı Bileşenin Boyutu')
    plt.title(title)
    plt.show()

    # En büyük bağlantılı bileşenin son hali
    final_largest_cc = largest_ccs[-1]
    print(f"En büyük bağlantılı bileşenin son hali: {final_largest_cc}")

plot_network_resilience(G)
```

```python
import networkx as nx
import community as community_louvain
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import numpy as np

# Louvain algoritması ile toplulukları bul
partition = community_louvain.best_partition(G)

# Toplulukları ve içerdikleri düğümleri yazdırma
for node, community in partition.items():
    print(f"Düğüm {node} topluluk {community}'a aittir.")

# Grafik düzenini ve renk paletini hazırlama
pos = nx.spring_layout(G)
colors = plt.cm.jet(np.linspace(0, 1, len(set(partition.values()))))

# Her topluluk için düğümleri farklı renkte çizme
for com, color in zip(set(partition.values()), colors):
    list_nodes = [nodes for nodes in partition.keys() if partition[nodes] == com]
    nx.draw_networkx_nodes(G, pos, list_nodes, node_size=20, node_color=[color])

# Kenarları çizme
nx.draw_networkx_edges(G, pos, alpha=0.5)

# Grafiği gösterme
plt.show()
```

```python
# Temel istatistikler
node_count = G.number_of_nodes()
edge_count = G.number_of_edges()
connected_components = nx.number_connected_components(G)
density = nx.density(G)
diameter = nx.diameter(G) if nx.is_connected(G) else "Graf bağlı değil"
avg_path_length = nx.average_shortest_path_length(G) if nx.is_connected(G) else "Graf bağlı değil"

# Derece dağılımı
degree_centrality = nx.degree_centrality(G)
closeness_centrality = nx.closeness_centrality(G)
betweenness_centrality = nx.betweenness_centrality(G)

# Kümelenme katsayısı
clustering_coefficient = nx.average_clustering(G)

# Yapısal benzerlik (Jaccard benzerlik katsayısı)
jaccard_similarity = list(nx.jaccard_coefficient(G))

# Sonuçların yazdırılması
print("Düğüm Sayısı:", node_count)
print("Kenar Sayısı:", edge_count)
print("Bağlı Bileşen Sayısı:", connected_components)
print("Yoğunluk:", density)
print("Çap:", diameter)
print("Ortalama Yol Uzunluğu:", avg_path_length)

print("\nDerece Dağılımı (Her Düğümün Derecesi):")
for node, degree in G.degree():
    print(f"{node}: {degree}")

print("\nDerece Merkeziliği (Her Düğüm İçin):")
for node, centrality in degree_centrality.items():
    print(f"{node}: {centrality:.3f}")

print("\nYakınlık Merkeziliği (Her Düğüm İçin):")
for node, centrality in closeness_centrality.items():
    print(f"{node}: {centrality:.3f}")

print("\nArasındalık Merkeziliği (Her Düğüm İçin):")
for node, centrality in betweenness_centrality.items():
    print(f"{node}: {centrality:.3f}")

print("\nOrtalama Kümelenme Katsayısı:", clustering_coefficient)

print("\nJaccard Benzerlik Katsayısı (İlk 5):")
for u, v, similarity in jaccard_similarity[:5]:
    print(f"({u}, {v}): {similarity:.3f}")
```
