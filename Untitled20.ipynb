```python
import networkx as nx
import random
import matplotlib.pyplot as plt

# Anz 'G' olarak ykleyin veya oluturun
# rnein: G = nx.gnp_random_graph(30, 0.05)

def plot_resilience(G, iterations=None):
    # Balangta an en byk balantl bileeninin boyutu
    initial_largest_cc = len(max(nx.connected_components(G), key=len))

    # Dm says
    num_nodes = G.number_of_nodes()

    # Varsaylan iterasyon says olarak dm saysn kullan
    if iterations is None:
        iterations = num_nodes - 1

    # Her dm karldnda en byk balantl bileenin boyutunu saklamak iin liste
    sizes = [initial_largest_cc]

    for i in range(iterations):
        # Rastgele bir dm se ve kar
        node_to_remove = random.choice(list(G.nodes()))
        G.remove_node(node_to_remove)

        # En byk balantl bileenin yeni boyutunu hesapla ve kaydet
        if len(G) > 0:  # Ada hala dm varsa
            largest_cc = len(max(nx.connected_components(G), key=len))
        else:
            largest_cc = 0
        sizes.append(largest_cc)

    # Yzdelik olarak boyutlar hesapla
    sizes_percent = [s / initial_largest_cc * 100 for s in sizes]

    # Grafii iz
    plt.plot(range(num_nodes - iterations, num_nodes + 1), sizes_percent)
    plt.xlabel('karlan Dm Says')
    plt.ylabel('En Byk Balantl Bileenin Boyutu (%)')
    plt.title('A Dayankll')
    plt.show()

# Fonksiyonu ararak a dayanklln iz
plot_resilience(G.copy())

```
